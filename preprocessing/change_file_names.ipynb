```python
import os

def rename_files_with_numbers(directory, prefix="data_", start_number=1):
    files = os.listdir(directory)
    
    for i, filename in enumerate(files, start=start_number):
        old_path = os.path.join(directory, filename)
        
        if os.path.isfile(old_path):
            file_extension = os.path.splitext(filename)[1]  # 파일 확장자
            new_filename = f"{prefix}{i}{file_extension}"
            new_path = os.path.join(directory, new_filename)
            
            os.rename(old_path, new_path)
            print(f"{old_path} -> {new_path}")

rename_files_with_numbers("data", prefix="data_", start_number=1)
```
