## 병합 및 배포

```python
data = data.reset_index(drop=True)
```

```python
result = pd.concat([data, col], axis=1)
```

```python
result = result[~((result['start'].apply(len) == 0) | 
                   (result['main'].apply(len) == 0) | 
                   (result['end'].apply(len) == 0))]
```

```python
result.drop(["MVM_PRSCRPTN_CN"], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

all_labels = [item for sublist in result["start"].tolist() + result["main"].tolist() + result["end"].tolist() for item in sublist]

encoder = LabelEncoder().fit(all_labels)
```

```python
result["start"] = result["start"].apply(encoder.transform)
result["main"] = result["main"].apply(encoder.transform)
result["end"] = result["end"].apply(encoder.transform)
```
